```python
movies = pd.read_csv('/content/drive/MyDrive/IMDB Dataset.csv')
```

```python

movies['review'] = movies['review'].str.lower()
```

```python
import re
def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'', text)
```

```python
movies['review'] = movies['review'].apply(remove_html_tags)
```

```python
import string,time
string.punctuation
exclude = string.punctuation

def remove_punc1(text):
    return text.translate(str.maketrans('', '', exclude))
```

```python
movies['review'] = movies['review'].apply(remove_punc1)
```

```python
movies.loc[0, 'review']
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]
    return " ".join(filtered_words)
```

```python
movies['review'] = movies['review'].apply(remove_stopwords)
```

```python
movies.tail()
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

# Function to lemmatize a single review
def lemmatize_review(review):
    words = word_tokenize(review)  # Tokenize the review into words
    lemmatized_words = [lemmatizer.lemmatize(word, 'v') for word in words]  # Lemmatize each word
    lemmatized_review = ' '.join(lemmatized_words)  # Join the lemmatized words back into a review
    return lemmatized_review

# Apply the lemmatization function to the 'review' column
movies['review'] = movies['review'].apply(lemmatize_review)
```

```python
movies.tail()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Initialize the CountVectorizer
count_vectorizer = CountVectorizer()

# Fit and transform the cleaned text data
count_vectorized = count_vectorizer.fit_transform(movies['review'])
print(count_vectorized)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize the TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the cleaned text data
tfidf_vectorized = tfidf_vectorizer.fit_transform(movies['review'])
print(tfidf_vectorized)
```

```python
from gensim.models import Word2Vec

# Split the cleaned text into word tokens
tokenized_texts = [text.split() for text in movies['review']]

# Train Word2Vec model
w2v_model = Word2Vec(sentences=tokenized_texts, vector_size=100, window=5, min_count=1, sg=0)

# Get the vector representation for a word
word_vector = w2v_model.wv['example']
print(word_vector)
```

```python
from gensim.models import KeyedVectors

# Load Google News Word2Vec model
google_news_model_path = "/content/drive/MyDrive/GoogleNews-vectors-negative300.bin"
google_news_model = KeyedVectors.load_word2vec_format(google_news_model_path, binary=True)

# Get the vector representation for a word
word_vector_google = google_news_model['example']
print(word_vector_google)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(count_vectorized, movies['review'], test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC

# Initialize and train SVC with CountVectorizer
svc_count = SVC()
svc_count.fit(X_train, y_train)
y_pred_svc_count = svc_count.predict(X_test)
accuracy_svc_count = accuracy_score(y_test, y_pred_svc_count)
print("Accuracy with SVC and CountVectorizer:", accuracy_svc_count)
```
